```python
import spacy
import pandas as pd

nlp = spacy.load("fr_core_news_md")
df['PostalAddress_'] = df['PostalAddress'].apply(nlp)
df['City_'] = df['City'].apply(nlp)


def match_addr(input_word, df):
    input_word = nlp(input_word)
    similarity = [fuzz.ratio(input_word,row) for row in df['PostalAddress_']]
    return df.iloc[similarity.index(max(similarity))]['PostalAddress_'].text

def match_city(input_word, df):
    input_word = nlp(input_word)
    similarity = [fuzz.ratio(input_word,row) for row in df['City_']]
   # similarity = [jaccard_score(input_word.vector, row.vector) for row in df['City_']]
    return df.iloc[similarity.index(max(similarity))]['City_'].text
```
